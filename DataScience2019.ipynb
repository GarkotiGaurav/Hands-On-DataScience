## Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3

print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label="Apple",marker = 'o')
plt.plot(x,y2,color="green",label='Kivi',linestyle = 'dashed')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruits price over time")
plt.legend()
plt.show()
```

## Scatter plot

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3

print(x)
print(y1)
print(y2)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y1,color='red',label="Apple",marker = '^')
plt.scatter(x,y2,color="green",label='Kivi',linestyle = 'dashed')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruits price over time")
plt.legend()
plt.show()
```

## Bar graph

```python
x_coordinate = np.array([0,1,2])*2
plt.bar(x_coordinate-0.25,[23,43,54],width = 0.5,label = 'Current Year',tick_label = ["Gold","Platimum","Silver"])
plt.bar(x_coordinate+0.25,[23,43,54],width = 0.5,label = 'Next Year')
plt.legend()
plt.xlabel("metal")
plt.ylabel("Price")

plt.show()
```

## Pie chart

```python
subjects = "Physics","Chemistry","Maths","Python"
weightes = [20,10,15,5]

plt.pie(weightes,labels=subjects,explode = (0.1,0,0,0),autopct="%1.1f%%",shadow=True)
plt.show()
```

### Multivariate Normal Distribution

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[1.6,0.4]])

dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)

print(dist)
```

```python
print(dist.shape)
plt.style.use("seaborn")
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])

plt.show()
```
